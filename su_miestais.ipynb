```python
years = miestai.columns[::-1]
print(f"metai: {years}")
clauses =[f"(miestai['{year}'] > miestai['{years[index+1]}'])" for index, year in enumerate(years[:-1])]
condition = "&".join(clauses)
print(condition)
miestai[eval(condition)]
```
