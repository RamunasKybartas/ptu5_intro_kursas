# grupavimas

```python
duomenys = {
    'salis':str('Lietuva,'*3+ 'Latvija,'*3+'Estija,'*3)[:-1].split(','),
    'miestai' : 'Vilnius,Kaunas,Klaipeda,RIga,Venspils,Daugpils,Talinn,Tartu,Parnu'.split(','),
    'gyventojai': [541, 287, 147, 716, 43, 105, 400, 101, 46] 
}
duomenys
miestai = pd.DataFrame(duomenys)
miestai
```

```python
str('Lietuva,'*3+ 'Latvija,'*3+'Estija,'*3).split(',')
```

```python
countries = miestai.groupby('salis')
countries.mean()
```

```python
countries.sum(numeric_only=True)
```

```python
countries.groups
```

```python
countries.groups.values()
```

```python
countries.count()
```

```python
countries.max()
```

```python
#konkrecios grupes duomenu traukimas
countries.sum(numeric_only=True).loc['Lietuva']
```

# DF jungimas

```python
mtop = miestai[:6]
```

```python
mbot = miestai[6:]
mbot
```

```python
pd.concat([mtop, mbot])
```

```python
pd.concat([mtop, mbot], axis=1)
```

```python
mright = mbot.reset_index()
mright.drop(columns='index', inplace=True)
mright
```

```python
pd.concat([mtop,mright], axis=1)
```

```python
lenkija = {
    'salis': ','.join(['Lenkija' for x in range(3)]).split(','),
    'miestai': 'Varsuva,Vroclavas,Gdanskas'.split(','),
    'gyventojai': [1688,638,461]
}
polonia = pd.DataFrame(lenkija)
polonia
```

```python
mpl = pd.concat([mright, polonia]).reset_index().drop(columns='index')
mpl
```

```python
pd.concat([mtop,mpl], axis=1)
```

```python
key = 'A,B,C,D,E,F'.split(',')
mtop['key'] = key
mpl['key'] = key
```

```python
pd.merge(mtop, mpl, on='key')
```

```python
key2 = 'B,C,D,E,F,G'.split(',')
mtop['key2'] = key2
mpl['key2'] = key2
```

```python
mtop, mpl
```

```python
pd.merge(mtop, mpl, on=['key', 'key2'])
```

```python
keyx = 'W,C,D,E,F,W'.split(',')
mpl['key2'] = keyx
mtop, mpl
```

```python
pd.merge(mtop, mpl, on=['key', 'key2'])
```
